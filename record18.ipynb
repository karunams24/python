```python
#Create a class Rectangle with private attributes length and width. Overload ‘<’ operator to
#compare the area of 2 rectangles. 
class rectangle:
    def __init__(self,len,wid):
        self.len=len
        self.wid=wid
        self.area=len*wid
        
    def __lt__(self,other):
        print("area of rectanlge1:",self.area)
        print("area of rectangle2:",other.area)
        if self.area<other.area:
            print("rectangle2 is greater")
        else:
            print("rectangle1 is greater")

#main
len1=int(input("enter the length of rect 1:"))
wid1=int(input("enter the width of rect 1:"))
len2=int(input("enter the length of rect 2:"))
wid2=int(input("enter the length of rect 2:"))

#objects
r1=rectangle(len1,wid1)
r2=rectangle(len2,wid2)
r1<r2

        
     
    
    
```
